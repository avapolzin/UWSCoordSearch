```python
%matplotlib inline
from uwstatus import search, plot
```

The observing status information is pulled from a regularly(-ish) updated file that's hosted on GitHub, so "last updated" below will reflect the last time that file was modified. There is no need to download it locally as it will continue to be updated as new observations occur.

# `uwstatus.search.status()`

Can be used to search for the status of a UW field, whether by searching that field directly or by searching a named object or set of coordinates that fall into the UW footprint.

```python
stat_frames, stat_obs = search.status('UW2449')
```

```python
obj_frames, obj_obs = search.status('NGC1052', type = 'name')
```

You can suppress the printed statement with `verbose = False`:

```python
coord_frames, coord_obs = search.status('10:00:30.03 +02:08:59.47', type = 'coord', verbose = False)
```

# `uwstatus.search.namesearch()`

Similar to `search.status()`, `search.namesearch()` returns the number of observed frames and the observed dates by relating the named object to a UW field, but it also outputs that field and does not require a `type = 'name'` argument.

```python
name_field, name_frames, name_obs = search.namesearch('NGC1052')
```

# `uwstatus.search.coordsearch()`

The same as `search.namesearch()`, but for coordinates (great for lesser known objects!).

```python
coord_field, coord_frames, coord_obs = search.coordsearch('02:41:46.8 -08:24:12')
```

# `uwstatus.plot.survey()`

```python
fig, ax = plot.survey()
```

You can also highlight various fields:

```python
plot.survey(highlight_fields = ['UW490', 'UW2400'])
```

# `uwstatus.plot.field()`

```python
plot.field('UW1775')
```

You can also show the virial radii of nearby galaxies (which also computes the coverage of the field) and set the crossmatch to occur only within a specific distance range.

```python
plot.field('UW1775', dist_limits = [19, 25], virial = True)
```

Much like you can highlight various fields in the `plot.survey()`, you can highlight coordinates in `plot.field()`:

```python
plot.field('UW1775', highlight_coords = ['02:41:46.8 -08:24:12'], verbose = False)
```
